Q1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" are treated as different words in case-sensitive processing, but they become the same when text is converted to lowercase; keeping case can preserve meaning, while lowercasing reduces complexity.

Q2. What happens if stopwords are not removed?

If stopwords are not removed, the text will contain more noise, increase computation, and reduce the importance of meaningful words.

Q3. Provide two real-world scenarios where removing stopwords can be harmful.

Removing stopwords can be harmful in tasks like sentiment analysis, where words like “not” change meaning, and in search or question answering, where stopwords help preserve the structure and intent of a sentence.

Q4. Explain the difference between stemming and lemmatization.

Stemming is a fast method that cuts words to their root form but may produce incorrect words, while lemmatization is a more accurate method that uses a dictionary to return the correct base form of a word.

# Task 2

```python
import re

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (e.g., sooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove very short tokens (≤2), except "no" and "not"
    words = text.split()
    filtered_words = [w for w in words if len(w) > 2 or w in ['no', 'not']]

    return ' '.join(filtered_words)
```

# Task 4

```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ['no', 'not']]

    return ' '.join(words)

sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Tracking for analysis
noise_scores = []
meaningful_scores = []

for i, text in enumerate(sentences, 1):
    cleaned = preprocess_text(text)
    tokens = cleaned.split()

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0

    # Noise = how much removed
    noise = len(text.split()) - total_tokens
    noise_scores.append((i, noise))

    # Meaningful = tokens retained
    meaningful_scores.append((i, total_tokens))

    print(f"\n--- {i} ---")
    print("Original Text:", text)
    print("Cleaned Sentence:", cleaned)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))

# Analysis
most_noise = max(noise_scores, key=lambda x: x[1])
most_meaningful = max(meaningful_scores, key=lambda x: x[1])

print("\n=== Analysis ===")
print(f"Sentence {most_noise[0]} had the most noise.")
print(f"Sentence {most_meaningful[0]} retained the most meaningful tokens.")
```

```python
from collections import Counter
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ['no', 'not']]

    return words  # return tokens directly

sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Combine all tokens
all_tokens = []

for text in sentences:
    tokens = preprocess_text(text)
    all_tokens.extend(tokens)

# Count frequencies
word_freq = Counter(all_tokens)

# Top 10 most frequent
top_10 = word_freq.most_common(10)

# Top 5 least frequent
least_5 = sorted(word_freq.items(), key=lambda x: x[1])[:5]

print("Top 10 Most Frequent Words:")
for word, freq in top_10:
    print(f"{word}: {freq}")

print("\nTop 5 Least Frequent Words:")
for word, freq in least_5:
    print(f"{word}: {freq}")
```

# Task 6

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(' '.join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

texts = [
    "Hello World!!!",
    "This is soooo goooood!!!"
]

result = full_pipeline(texts)

print(result)
```
